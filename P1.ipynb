## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
def make_region_interest(img):
    """
    Calculate range of interest from image shape
    
    Returns (left_bottom, right_bottom, left_top, right_top)
    """

    H, W = get_img_size(img)
    
    left_bottom = [0.05*W, H-1]
    right_bottom = [0.95*W, H-1]
    left_top = [0.40*W, 0.65*H]
    right_top = [0.6*W, 0.65*H]
    
    return np.array([[left_bottom, right_bottom, right_top, left_top]], dtype=np.int32)
    
def get_hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    return lines

def get_averaged_lines(img, lines, theta_cutoff = 0):
    """
    Averaging lines with thier ends points by using fitting
    
    Return two lines which correspond to left and right lanes
    """
    H, W = get_img_size(img)
        
    left_bottom, right_bottom, right_top, left_top = make_region_interest(img).squeeze()
    y_bottom = left_bottom[1]
    y_top = left_top[1]
    
    x_left = []
    y_left = []
    w_left = []
    
    x_right = []
    y_right = []
    w_right = []
    
    for line in lines.squeeze():
        theta, d = get_angle_length(line)
        if (theta is None) or (abs(theta) < theta_cutoff):
            continue
        x1, y1, x2, y2 = line
        if (x1 + x2)/2 >= W/2: 
            x_right += [x1, x2]
            y_right += [y1, y2]
            w_right += [d, d]
        else:
            x_left += [x1, x2]
            y_left += [y1, y2]
            w_left += [d, d]            
    line_left = get_averaged_line(x_left, y_left, w_left, y_bottom, y_top)
    line_right = get_averaged_line(x_right, y_right, w_right, y_bottom, y_top)

    return [[line_left], [line_right]]


def get_averaged_line(x, y, w, y1, y2):
    """
    Averaging lines with thier ends points by using fitting
    
    Return an averaged line
    """
    p = np.polyfit(x, y, 1, w=w)
    
    x1 = int((1/p[0]) * y1 - p[1]/p[0])
    x2 = int((1/p[0]) * y2 - p[1]/p[0])

    return (x1, y1, x2, y2)

def get_img_size(img):
    
    if len(img.shape) == 2:
        H, W = img.shape
    else:
        H, W, _ = img.shape
    
    return H, W

def get_angle_length(line):
    """
    Calculate angle of line from the axis and length of line
    `line` is flatten pairs of points (x1, y1, x2, y2)

    Returns (angle from the x axis, length of  the line)
    """
    x1, y1, x2, y2 = line

    if y2-y1 == 0:
        return (0, abs(x2-x1))

    d = np.linalg.norm(np.array([x2, y2]) - np.array([x1, y1]))
    
    if x2-x1 == 0:
        theta = np.sign(y2-y1)*np.pi/2    
    else:
        theta = np.arctan((y2-y1)/(x2-x1))
    return (theta, d)


def get_time_average(img, line_history, decay_rate=0.9):
    """
    Calculate time average of `line_history`
    
    Return time averaed line
    """
    
    left_bottom, right_bottom, right_top, left_top = make_region_interest(img).squeeze()
    y_bottom = left_bottom[1]
    y_top = left_top[1]
    
    x = []
    y = []
    w = []
    r = decay_rate
    
    for i, line in enumerate(line_history):
        x1, y1, x2, y2 = line
        x += [x1, x2]
        y += [y1, y2]
        w += [r**i, r**i]
    
    return get_averaged_line(x, y, w, y_bottom, y_top)

    
````

````python
def show_images(images, names):
    plt.figure(figsize=(17, 18))
    for i in range(len(images)):
        ax = plt.subplot(3, 2, i+1)
        ax.set(title=names[i])
        cmap = 'gray' if len(images[i].shape) == 2 else None
        plt.imshow(images[i], cmap=cmap)

def save_image(image, fname, outdir='./test_images_out' ):
    plt.figure(figsize=(4, 3))
    cmap = 'gray' if len(image.shape) == 2 else None
    plt.imshow(image, cmap=cmap)
    plt.savefig("{0}/{1}".format(outdir, fname), bbox_inches='tight')
````

````python
from collections import deque

hist_len = 15
left_line_history = deque(maxlen=hist_len)
right_line_history = deque(maxlen=hist_len)

def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    
    gaussian_kernel_size = 5
    canny_thresh_low = 100
    canny_thresh_high = 200

    rho = 5
    theta =1*np.pi/180
    hough_threshold = 15
    min_line_length = 15
    max_line_gap = 7
    theta_cutoff = np.pi/10
    decay_rate = 0.9
    
    ## apply gaussian smoothing
    gaussian_blur(grayscale(image), gaussian_kernel_size)
    ## apply canny edge detection 
    img_canny = canny(image, canny_thresh_low, canny_thresh_high)
    ## extract 
    mask_region = make_region_interest(img_canny)
    ## show extract region as blue area
    blue_mask = np.zeros_like(image) + np.array([0, 0, 200], dtype=np.uint8)
    ## calculate hough transformed line segments
    h_lines = get_hough_lines(region_of_interest(img_canny, mask_region),
                              rho, theta, hough_threshold, min_line_length, max_line_gap)

    ## calculate average of line segments. theta_cutoff is used to remove noisy horizontal segment which distracts averaged line
    avg_lines = get_averaged_lines(img_canny, h_lines, theta_cutoff)

    ## store line history for time averaging
    left_line_history.appendleft(avg_lines[0][0])
    right_line_history.appendleft(avg_lines[1][0])
    ## calculate time average of line
    time_avg_line_left = get_time_average(image, left_line_history, decay_rate)
    time_avg_line_right = get_time_average(image, right_line_history, decay_rate)    
    time_avg_lines = [[time_avg_line_left], [time_avg_line_right]]
    ## create new image data to draw line 
    img_line = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
    ## draw hough transformed line segments as information for debug or reference
    draw_lines(img_line, h_lines)
    ## draw lines which are time averaged lines.
    draw_lines(img_line, time_avg_lines, color=[200, 200, 0], thickness=4)     

    result = weighted_img(img_line, image)
    
    result = weighted_img(region_of_interest(blue_mask, mask_region), result)
    
    return result
````
